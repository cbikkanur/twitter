```python
import twitter
import pandas as pd
import json
```

```python
with open("cbikkanur_twitter_credentials.json", "r") as file:
    credentials = json.load(file)
```

```python
api = twitter.Api(consumer_key=credentials["consumer_key"], \
                  consumer_secret=credentials["consumer_secret"], \
                  access_token_key=credentials["access_token_key"], \
                  access_token_secret=credentials["access_token_secret"], \
                  tweet_mode='extended')
```

```python
users = api.GetFriends()
print([[u.name, u.id] for u in users])
```

```python
user_ids = api.GetFriendIDs()
print([u for u in user_ids])
```

```python
def getTweetsFromTimeline(user_ids, count=10):
    list_tweets = []
    for user_id in user_ids:
        tweets = api.GetUserTimeline(user_id, count = count)
        for tweet in tweets:            
            list_tweets.append([tweet.id, tweet.user.id, tweet.user.name, tweet.user.screen_name, 'Bank/Financial', tweet.full_text])        
    df = pd.DataFrame(list_tweets, columns =['Tweet_Id', 'User_Id', 'User_Name', 'User_Screen_Name', 'Theme','Text'],)
    return df
    
```

```python
df1 = getTweetsFromTimeline(user_ids, count=200) 
df1
```

```python
def getTweetsFromSearch(search_words, since, count=10):
    list_tweets = []
    for search_word in search_words:
        tweets = api.GetSearch(term=search_word, lang = 'en', since=since, count=count)
        for tweet in tweets:            
            list_tweets.append([tweet.id, tweet.user.id, tweet.user.name, tweet.user.screen_name, search_word, tweet.full_text])        
    df = pd.DataFrame(list_tweets, columns =['Tweet_Id', 'User_Id', 'User_Name', 'User_Screen_Name','Theme', 'Text'],)
    return df
    
```

```python
search_words = ["@Citi", "covid-19", "cake", "dog","car", "nasa", "tokyo", "airlines", "ramadan", "coffee"]
df2 = getTweetsFromSearch(search_words, since = 2020-4-23, count=100)
df2
```

```python
df3 = df1.append(df2, ignore_index=True)
df3.tail(10)
```
